```python
import pandas as pd

# Load the dataset
file_path = '../../data/WSJ/WSJ_data_modified.csv'
data = pd.read_csv(file_path)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Correcting the issue by ensuring all entries are treated as strings, even if they are missing (NaN)
text = ' '.join(data['Title'].fillna('') + ' ' + data['Summary'].fillna(''))

# Re-generate the word cloud with corrected data
wordcloud = WordCloud(width = 800, height = 400, background_color ='white').generate(text)

# Display the corrected word cloud image
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
