```python
import numpy as np

def simulate_game(exit_after):
    score = 2  # 初始积分
    time = 0  # 初始化时间
    n = 0  # 敌机击落数初始化
    while True:
        n += 1
        # 生成下一架敌机出现的时间（指数分布）
        interarrival_time = np.random.exponential(1)
        time += interarrival_time
        # 更新积分（时间消耗）
        score -= interarrival_time
        # 检查积分是否已经用尽
        if score <= 0:
            return 0
        # 计算是否被击落
        p_success = 0.85 ** n
        if np.random.random() > p_success:  # 被击落的情况
            return score
        # 击落成功，积分增加
        score += 1.5
        # 检查是否到达退出点
        if n == exit_after:
            return score

# 设置模拟次数
simulations = 1000000



# 对于每个退出点（1到4架敌机），进行模拟
results = {n: np.mean([simulate_game(n) for _ in range(simulations)]) for n in [1,2,3,4]}

results
```

```python
from scipy.stats import binom
import numpy as np

# Given parameters
S0 = 10          # Spot price
u = 1.1          # Up-factor
d = 0.9          # Down-factor
r = 0.05         # Risk-free interest rate
K = 10.5         # Strike price
T = 0.5          # Time to maturity in years
N = int(T / 0.1) # Number of time steps
delta_t = 0.1    # Time step in years

# Calculate the risk-neutral probability
q_star = (np.exp(r * delta_t) - d) / (u - d)

# Discount factor
discount_factor = np.exp(-r * T)

# Initialize the option price
C = 0

# Calculate the option price using the binomial model
for j in range(N + 1):
    S_T = S0 * (u ** j) * (d ** (N - j))  # Stock price at time T for j up-moves
    payoff = max(S_T - K, 0)              # Payoff of the call option
    probability = binom.pmf(j, N, q_star) # Binomial probability of j up-moves
    C += payoff * probability

# Discount the expected payoff to present value
C *= discount_factor

C
```
